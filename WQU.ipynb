```python
import numpy as np
import pandas as pd
import numpy_financial as npf
```

```python
Table1=pd.DataFrame(columns=['Month','Fixed_Payment_Amount','Principal_Payment','Interest_Payment','Final_Principal'])
```

```python
def table(principal,ir, total_year, grid):
    payment= -(npf.pmt(ir / 12, total_year * 12,  principal, 0, when='end'))
    grid.at[0,'Final_Principal']=principal
    grid.at[0,'Month']=0
    for i in range(1,(total_year*12)+1):
        payment=payment
        rate=grid['Final_Principal'][i-1]*ir/12
        princip=payment-rate
        final=grid['Final_Principal'][i-1]-princip
        grid = grid.append({'Month': i, 'Fixed_Payment_Amount': payment, 'Principal_Payment': princip, 'Interest_Payment': rate, 'Final_Principal': final},ignore_index=True)
        #new=pd.DataFrame(grid)
    return grid
        
    
```

```python
sim_rates=sim_rates.set_index('SNO')
```

```python
#Creation of a function that produces the ARM TABLE with input of Principal Balance, Introductory Interest Rate, and Initial Period of low rate

def arm(principal,ir, initial,total_year, grid):
    payment= -(npf.pmt(ir / 12, total_year * 12,  principal, 0, when='end'))
    grid.at[0,'Final_Principal']=principal
    grid.at[0,'Month']=0
    for i in range(1,(initial*12)+1):
        payment=payment
        rate=grid['Final_Principal'][i-1]*ir/12
        princip=payment-rate
        final=grid['Final_Principal'][i-1]-princip
        grid = grid.append({'Month': i, 'Fixed_Payment_Amount': payment, 'Principal_Payment': princip, 'Interest_Payment': rate, 'Final_Principal': final},ignore_index=True)   
    
    start=1476 #starting location of interest rate from the MORTGAGE30US Table
       
    for x in range((initial*12)+1,(total_year*12)+1,6):
        n_rate=(sim_rates.loc[start]['MORTGAGE30US'])/100
        n_payment=-(npf.pmt(n_rate / 12, (total_year * 12)-(x-1),  grid.iloc[(x-1)]['Final_Principal'], 0, when='end'))
        n=x
        for j in range(1,7):
            payment=n_payment
            rate=grid['Final_Principal'][n-1]*n_rate/12
            princip=payment-rate
            final=grid['Final_Principal'][n-1]-princip
            grid = grid.append({'Month': n, 'Fixed_Payment_Amount': payment, 'Principal_Payment': princip, 'Interest_Payment': rate, 'Final_Principal': final},ignore_index=True)
            n+=1
        start+=26
        
    return grid
        
    
```

```python
Table_Third=arm(1000000,0.035,7,30,Table1)
```

```python
Total_interest_third = Table_Third['Interest_Payment'].sum()
print("The total interest is : ",Total_interest_third)
```
